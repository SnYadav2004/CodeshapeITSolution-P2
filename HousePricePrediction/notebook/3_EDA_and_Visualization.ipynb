# ✅ Step 1: Setup & Load Data

```python
# Load required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore warnings
warnings.filterwarnings('ignore')
```

# import data from file

```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv('cleaned_house_data.csv')

# Preview the data
df.head()
```

# import from sql

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:8005130826@localhost/house_price')
df = pd.read_sql('SELECT * FROM house_data', con=engine)
```

```python
df.head()
```

```python
# Shape of dataset
print("Rows, Columns:", df.shape)

# Info about columns
df.info()

# Statistical summary
df.describe()
```

```python
# Histogram of SalePrice
plt.figure(figsize=(8, 5))
sns.histplot(df['SalePrice'], kde=True, bins=30)
plt.title('SalePrice Distribution')
plt.show()
```

# ✅ Step 4: Correlation Heatmap

```python
# Correlation matrix
correlation = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

# ✅ Step 5: Top 10 Features Correlated with SalePrice

```python
# Correlation with target
top_corr = correlation['SalePrice'].abs().sort_values(ascending=False).head(10)

print("Top correlated features:\n", top_corr)

# Plot them
plt.figure(figsize=(10, 6))
top_corr.drop('SalePrice').plot(kind='barh')
plt.title('Top Features Correlated with SalePrice')
plt.gca().invert_yaxis()
plt.show()
```

# ✅ Step 6: Visualize Key Relationships

```python
# Example 1: OverallQual vs SalePrice
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title('SalePrice vs OverallQual')
plt.show()

# Example 2: GrLivArea vs SalePrice
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('SalePrice vs GrLivArea')
plt.show()

# Example 3: YearBuilt vs SalePrice
sns.scatterplot(x='YearBuilt', y='SalePrice', data=df)
plt.title('SalePrice vs YearBuilt')
plt.show()
```
